```python
cap = cap.rename(columns = {'Area_Name':'state','CPA_-_Complaints_Received/Alleged':'Alleged'})
```

```python
cap.groupby('state').agg({'Alleged': sum}).sort_values('Alleged',ascending = False).head(10).plot.bar()
```

```python
bar = cap.groupby('state').agg({'Alleged': sum}).sort_values('Alleged',ascending = False).reset_index().head(10)
plt.figure(figsize=(25,15))
sns.barplot(bar.state,bar.Alleged)
plt.rc('font', size=20)
#plt.rcParams.update({'font.size': 22})
#plt.text(fonsize=22))
```
